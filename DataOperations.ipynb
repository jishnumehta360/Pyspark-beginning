```python
import pyspark as py
import pandas as pd
import numpy as np
!apt-get install openjdk-8-jdk-headless -qq > /dev/null
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Operations").getOrCreate()
```

```python
spark
```

```python
#Reading the schema with appropriate datatypes
df_spark = spark.read.option('header','true').csv('/content/sample_data/test1.csv',inferSchema=True)
df_spark.printSchema()
```

```python
df_spark_1 = spark.read.csv('/content/sample_data/test1.csv',header=True,inferSchema=True)
df_spark_1.show()
df_spark_1.columns
```

```python
df_spark_1.select('Name','Experience').show()
df_spark_1.dtypes
df_spark_1.describe()
```

```python
##Adding Columns

df_spark_2 = df_spark_1.withColumn('Experience After 2 years',df_spark_1['Experience']+2)
df_spark_2.show()
```

```python
##Dropping Columns

df_spark_3 = df_spark_2.drop('Experience After 2 years')
df_spark_3.show()
```

```python
##Renameing Columns
df_spark_4 = df_spark_3.withColumnRenamed('Name','New Name')
df_spark_4.show()
```

```python

##Dropping null values
spark_2 = spark.Builder().appName("Practice").getOrCreate()
df_spark_6 = spark_2.read.csv('/content/sample_data/test2.csv',header=True,inferSchema=True)

df_spark_6.na.drop().show()
```

```python
df_spark_6.na.drop(how='any',thresh=2).show()
df_spark_6.na.drop(how='all').show()
df_spark_6.show()
```

```python
df_spark_6.na.drop(how='any', subset=['Experience']).show()
```

```python
df_spark_6.printSchema()
df_spark_7 = spark_2.read.csv('/content/sample_data/test2.csv',header=True)

## Only replaces null values when column type is String

df_spark_7.na.fill('Missing Values', ['Name','age','Experience']).show()
df_spark_7.na.fill('Missing Values', ['age', 'Experience']).show()
```

```python
## Imputer function

from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'Experience', 'Salary'],
    outputCols=["{}_imputed".format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy("mean")
```

```python
imputer.fit(df_spark_6).transform(df_spark_6).show()
```

```python
## Filtering Operations
df_spark_6.filter("Salary<=30000").show()
df_spark_6.filter("Salary<=30000").select('Name','age','Salary').show()
df_spark_6.filter(~(df_spark_6['Salary']>=20000)).show()
df_spark_6.filter((df_spark_6['Salary']<=20000) | (df_spark_6['Salary'] >=35000)).show()
```
